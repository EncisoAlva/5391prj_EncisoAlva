```python
import requests
import pandas as pd
```

```python
def get_crypto_price(symbol, exchange, start_date = None):
    api_key = 'IBU64LG9CX3F1RQ6'
    api_url = f'https://www.alphavantage.co/query?function=DIGITAL_CURRENCY_DAILY&symbol={symbol}&market={exchange}&apikey={api_key}'
    raw_df = requests.get(api_url).json()
    df = pd.DataFrame(raw_df['Time Series (Digital Currency Daily)']).T
    df = df.rename(columns = {'1a. open (USD)': 'open', '2a. high (USD)': 'high', '3a. low (USD)': 'low', '4a. close (USD)': 'close', '5. volume': 'volume'})
    df = df.iloc[::-1].drop(['1b. open (USD)', '2b. high (USD)', '3b. low (USD)', '4b. close (USD)', '6. market cap (USD)'], axis = 1)
    for i in df.columns:
        df[i] = df[i].astype(float)
    df.index = pd.to_datetime(df.index)
    if start_date:
        df = df[df.index >= start_date]
    return df

btc = get_crypto_price( symbol='BTC', exchange='USD', start_date=None)
#btc
```

```python
def get_crypto_data_USD(symbol, start_date=None):
    api_key = 'IBU64LG9CX3F1RQ6'
    api_url = f'https://www.alphavantage.co/query?function=DIGITAL_CURRENCY_DAILY&symbol={symbol}&market=USD&apikey={api_key}'
    raw_df = requests.get(api_url).json()
    df = pd.DataFrame(raw_df['Time Series (Digital Currency Daily)']).T
    #print(tabulate(df,tablefmt='psql'))
    df = df.rename(columns = {'2a. high (USD)': f'{symbol}_max', '3a. low (USD)': f'{symbol}_min', '5. volume': f'{symbol}_vol'})
    df = df.iloc[::-1].drop(['1a. open (USD)','1b. open (USD)', '2b. high (USD)', '3b. low (USD)', '4a. close (USD)', '4b. close (USD)', '6. market cap (USD)'], axis = 1)
    for i in df.columns:
        df[i] = df[i].astype(float)
    df.index = pd.to_datetime(df.index)
    if start_date:
        df = df[df.index >= start_date]
    return df

btc = get_crypto_data_USD( symbol='BTC', start_date=None)
#btc
```

```python
crypto = dat.to_numpy()
plt.plot(crypto[:,0])
plt.show()
```

```python
dmd = DMD(svd_rank=4)
dmd.fit(crypto)
```

```python
for eig in dmd.eigs:
    print('Eigenvalue {}: distance from unit circle {}'.format(eig, np.abs(eig.imag**2+eig.real**2 - 1)))
dmd.plot_eigs(show_axes=True, show_unit_circle=True)
```

```python
for mode in dmd.modes.T:
    plt.plot(mode.real)
    plt.title('Modes')
plt.show()
for dynamic in dmd.dynamics:
    plt.plot(dynamic.real)
    plt.title('Dynamics')
plt.show()
```
